```python
#Before going too much further into the analysis, I want to combine the datasets into one big dataframe

all_df_list = [Amsterdam_df, Belgium_df, Athens_df, Austin_df, Bangkok_df, Barcelona_df, Berlin_df, CapeTown_df, Dublin_df, HongKong_df, NewYork_df, Australia_df] 

combined_dfs = pd.concat(all_df_list, ignore_index=True) 
```

## Inspecting & Dealing with Missing Data

In the planning stages dropping nans was the descision. Upon exploring missing data, keeping everything intact and filtering per data requests will be the method moving forward with this project. 

```python
print(combined_dfs.isna().sum())
```

```python
print(combined_dfs.shape)

for col in [
    'name',
    'host_name',
    'neighbourhood_group',
    'price',
    'last_review',
    'reviews_per_month',
    'license',
    'host_profile_id'
]:
    temp = combined_dfs.dropna(subset=[col])
    print(f"{col}: {temp.shape}")

#Essentially dropping nans of one column will alter the data heavily. Therefore leaving everything as is, nans are still useful to an extent.
```

## Exploratory Data Analysis & Summary Statistics

I'm only going to investigate and collect data on relevant columns:
** room_type
** number_of_reviews_ltm
** price
** reviews_per_month

```python
print(combined_dfs.room_type.head())

print(combined_dfs.room_type.isna().sum())

print(combined_dfs.room_type.value_counts()) #Counts

print(combined_dfs.room_type.value_counts(normalize=True)) #Percentages
```

```python
sns.countplot(x='room_type', data=combined_dfs)
plt.show()
plt.clf()
```

```python
print(combined_dfs.number_of_reviews_ltm.head())

print(combined_dfs.number_of_reviews_ltm.isna().sum())

# Central Tendancy

print(round(combined_dfs.number_of_reviews_ltm.mean()),) # mean = 9
print(round(combined_dfs.number_of_reviews_ltm.median()),) # median = 2
print(round(combined_dfs.number_of_reviews_ltm.mode()),) # mode = 0

# Spread

reviews_ltm_range = (combined_dfs.number_of_reviews_ltm.max() - combined_dfs.number_of_reviews_ltm.min())
print(reviews_ltm_range) # range = 1304
reviews_ltm_iqr = (combined_dfs.number_of_reviews_ltm.quantile(0.75) - combined_dfs.number_of_reviews_ltm.quantile(0.25))
print(reviews_ltm_iqr) # iqr = 10
print(combined_dfs.number_of_reviews_ltm.var()) # variance = 355.77
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='number_of_reviews_ltm', data=combined_dfs)
plt.show()
plt.clf()
```

## Creating Data Visualizations

```python
active_listings = combined_dfs[combined_dfs['availability_365'] > 0]

print(combined_dfs.shape)
print(active_listings.shape)

pct_removed = (
    (len(combined_dfs) - len(active_listings))
    /len(combined_dfs)
    ) * 100

print(f"{pct_removed:.2f}% removed from dataframe")
```

```python
demand_by_room = (
    active_listings.groupby('room_type')['number_of_reviews_ltm']
    .median()
    .sort_values(ascending=False)
)

demand_by_room.plot(kind='bar') #Keep in mind that shared rooms only has 955 rows in our column 
plt.title("Property Type Demand")
plt.ylabel("Median Reviews Per LTM (Last Twelve Months)")
plt.xlabel("Airbnb Type")
plt.xticks(size=8, rotation=45)

plt.savefig('property-demand.png', dpi=300, bbox_inches='tight')

plt.show()
plt.clf()

```

```python
active_listings['price'].describe()

active_listings['price_bucket'] = pd.qcut(
    active_listings['price'],
    q=4, 
    labels= [
        'Low price',
        'Mid-low price',
        'Mid-high price',
        'High price'
    ]
)

price_demand = (
      active_listings.groupby('price_bucket')['number_of_reviews_ltm']
    .median()
)

print(price_demand)
```

```python
price_demand.plot(kind='bar', color='green') 
plt.title("Price Demand")
plt.ylabel("Median Reviews per LTM (Last Tweleve Months)")
plt.xlabel("Price Ranges")
plt.xticks(size=8, rotation=45)

plt.savefig('price-demand.png', dpi=300, bbox_inches='tight')

plt.show()
plt.clf()
```

```python
active_listings['minimum_nights'].describe()

night_bins = [0, 1, 3, 7, 30, float('inf')]

night_labels = [
    '1 Night',
    '2-3 Nights',
    '4-7 Nights',
    '8-30 Nights',
    '30+ Nights']

active_listings['night_bucket'] = pd.cut(
    active_listings['minimum_nights'],
    bins=night_bins,
    labels=night_labels
)

nights_demand = (
      active_listings.groupby('night_bucket')['number_of_reviews_ltm']
    .median()
)

print(nights_demand)
```

```python
nights_demand.plot(kind='bar')
plt.title('Night Demand')
plt.ylabel('Median Reviews per LTM (Last Tweleve Months)')
plt.xlabel('Minimum Nights')
plt.xticks(size=8, rotation=45)

plt.savefig('night-demand.png', dpi=300, bbox_inches='tight')

plt.show()
plt.clf()
```

```python
active_listings['number_of_reviews'].describe()

pricing_influence = (
      active_listings.groupby('price_bucket')['number_of_reviews']
    .median()
)

print(pricing_influence)
```

```python
pricing_influence.plot(kind='bar', color='green')
plt.title('Price and Occupancy')
plt.ylabel('Median Number Of Reviews')
plt.xlabel('Price Quartiles')
plt.xticks(size=8, rotation=45)

plt.savefig('price-occupancy.png', dpi=300, bbox_inches='tight')

plt.show()
plt.clf()
```
